```python
import requests
import json
```

```python
from config import g_key
```

```python
city = "White House"

# params = {"address":city,"key":g_key}
```

```python
# Base google url
url = (f"https://maps.googleapis.com/maps/api/geocode/json?address={city}&key={g_key}")



response = requests.get(url)

geo = response.json()
geo
```

```python
# Geocodes of Seattle, Washington
WH_lat =geo["results"][0]["geometry"]["location"]["lat"]
```

```python
WH_lng=geo["results"][0]["geometry"]["location"]["lng"]
```

```python
#Build new code for next search
url_bikestore_dc=(f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?key={g_key}&location={WH_lat},{WH_lng}&radius=25000&name=cycle")
url_bikestore_dc

bike_stores=requests.get(url_bikestore_dc)
bike_stores
bike_stores_dc=bike_stores.json()
bike_stores_dc
                       
                       
```
